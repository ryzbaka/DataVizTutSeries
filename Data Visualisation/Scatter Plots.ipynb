For this module, the dataset that we'll be using is one of **Insurance Charges** i.e. how much do people pay for health insurance and what are their characteristics.

### Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
oo=pd.read_csv('insurance.csv')
```

```python
oo.head()
```

#### Simple Scatterplot

```python
sns.scatterplot(x=oo.bmi,y=oo.charges)
```

The scatterplot above suggests that body mass index (BMI) and insurance charges are positively correlated, where customers with higher BMI typically also tend to pay more in insurance costs. (This pattern makes sense, since high BMI is typically associated with higher risk of chronic disease.)

But how strong is this relationship?

To figure this out we use **linear regression**.

A linear regression is basically a line that *best fits* the data i.e. approximates a linear relationship.


#### Regression Plot

We can use a regression plot to draw this line of best fit.

```python
sns.regplot(x=oo.bmi,y=oo.charges)
```

As we can see, the slope of the linear regression line is not alot.
So while there may be a positive correlation between a person's BMI and the amount they pay for health insurance, it isn't a very strong relation.

So we might have to add in another characteristic in to the mix to see if it has a strong corellation with the amount paid for health insurance.

#### Simple Scatter Plot with Hue

We can see in the data that there's a column that tells is if a person is a smoker or not.

But how exactly can we represent this third feature on a graph( feature is the same as column and instance is the same as row, its data science jargon so get used to it).

We can use **color** to represent this feature.

```python
sns.scatterplot(x=oo.bmi,y=oo.charges,hue=oo.smoker)
```

The **hue** parameter is reponsible for coloring the data points based on the value of a column.

Now, we can see that people who smoke are more likely to pay higher health insurance and that this is a pretty significant relationship.

Lets see how to plot multiple linear regression lines for each hue to verify out inferences.

#### Multiple Regression Plot

In the code below, pay close attention to the parameters.

```python
sns.lmplot(x="bmi",y="charges",hue="smoker",data=oo)
```

This graph verifies our inference that people with high BMI who smoke pay more health insurance.

#### Swarm Plot

The last scatter plot that we'll be covering in this module is the **Swarm Plot**, which classifies data points on some value and shows its distribution.

```python
sns.swarmplot(x=oo.smoker,y=oo.charges)
```

Notice that the x value is the classification feature('smoker or non smoker) and the y value is the feature (insurance charge) who's distribution we want to see.

We can infer that most non smokers pay less than **20,000 USD** in health insurance while most smokers pay **30,000 to 50,000**


```python
sns.swarmplot(x=oo.charges,y=oo.smoker)
```
