In this module we'll be using the Iris flowers dataset which contains information about three flower species:-

1)<a href="https://en.wikipedia.org/wiki/Iris_virginica">Iris Virginica</a>

2)<a href="https://en.wikipedia.org/wiki/Iris_setosa">Iris Setosa</a>

3)<a href="https://en.wikipedia.org/wiki/Iris_versicolor">Iris Versicolor</a>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
oo=pd.read_csv('iris.csv',index_col='Id')
```

```python
oo.head()
```

### Histograms

Histograms help us understand the distribution of values of a feature in a dataset.

If we want to see the distribution of the length of Iris flowers, then we can plot the following histogram. 

```python
sns.distplot(a=oo['Petal Length (cm)'])
```

The smooth wavy line there is called a kernel density estimate and is a sort of smoothened histogram that gives the idea of distribution density at different values.

You can turn it off like this:-

```python
sns.distplot(a=oo['Petal Length (cm)'],kde=False)
```

### Density Plots

Suppose we just want the KDE of the Iris flower's petal width.

We can do so like this:-

```python
sns.kdeplot(data=oo['Petal Width (cm)'])
```

If you want to fill in the area under the curve then you can add a **shade=True** parameter like this:-

```python
sns.kdeplot(data=oo['Petal Width (cm)'],shade=True)
```

### 2D Distribution Plots

So far, we have only seen the distribution of plots of a single feature.

We can use 2D KDE plots to see how likely it is to encounter a combination of 2 features(like a combination of a Petal Length and Sepal Width) 

```python

sns.jointplot(x=oo['Petal Length (cm)'],y=oo['Sepal Width (cm)'],kind='kde',color='purple')
```

From this graph we can see that the darker shaded regions indicate the most common combination of the two feature values in the dataset.

 *Note: You can change the color of any Seaborn plot by just adding a **color=** parameter followed by a string that contains the color's name. You can see the list of compatible colors and color palettes <a href="https://python-graph-gallery.com/python-colors/">here</a>.*

In the 2D KDE plot, the curve at the top is a kde plot of the x axis and the one on the right is a kde plot of the y axis.

### Color-coded Plots

Lets try plotting the features of each of the three Iris Species.

```python

iris_set_data = pd.read_csv('iris_setosa.csv', index_col="Id")
iris_ver_data = pd.read_csv('iris_versicolor.csv', index_col="Id")
iris_vir_data = pd.read_csv('iris_virginica.csv', index_col="Id")

```

```python
sns.distplot(a=iris_set_data['Petal Length (cm)'],label='Iris-Setosa',kde=False)
sns.distplot(a=iris_ver_data['Petal Length (cm)'],label='Iris-Versicolor',kde=False)
sns.distplot(a=iris_vir_data['Petal Length (cm)'],label='Iris-Virginica',kde=False)
plt.legend()
```

The label tag is for the plot's legend.

```python
sns.kdeplot(data=iris_set_data['Petal Length (cm)'], label="Iris-setosa", shade=True)
sns.kdeplot(data=iris_ver_data['Petal Length (cm)'], label="Iris-versicolor", shade=True)
sns.kdeplot(data=iris_vir_data['Petal Length (cm)'], label="Iris-virginica", shade=True)
```
