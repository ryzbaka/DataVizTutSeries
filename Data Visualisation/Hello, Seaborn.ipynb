#### Import Packages

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Loading Data from a .csv file

```python
oo=pd.read_csv('fifa.csv',index_col='Date',parse_dates=True)
```

```python
type(list(oo.index)[0])
```

```python
oo.head()
```

Now that we have our data, lets use a *Line Plot* to visualise this data

```python
plt.figure(figsize=(20,10))# adjust this figure size,you can change the tuple values to plat around with this feauture
sns.lineplot(data=oo)
```

So, as we can see that the line plot shows us the rank of various companies in FIFA at various dates.
